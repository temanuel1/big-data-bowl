```python
import pandas as pd
```

```python
data = pd.read_csv('/Users/tyleremanuel/Bench/BDB Code-Data/clean_tackles.csv')
```

```python
def find_closest_tackler(game_id, play_id):
    play_data = pd.DataFrame()
    for index, row in data.iterrows():
        if row['gameId'] == game_id[0] and row['playId'] == play_id[0]:
            play_data[index] = row
    return play_data
```

```python
test_data = data.iloc[:1][['displayName', 'x', 'y', 'gameId', 'playId', 'club']]
print(test_data)
```

```python
find_closest_tackler(test_data['gameId'], test_data['playId'])
```
